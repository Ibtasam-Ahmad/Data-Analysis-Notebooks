**Let's consider a real network.**   
Let's read the file "Cross_Parker-Consulting_info.txt" that you will find in the folder "data"

The ties are differentiated on a scale from 0 to 5 in terms of frequency of information or advice requests (“Please indicate how often you have turned to this person for information or advice on work-related topics in the past three months”). 0: I Do Not Know This Person; 1: Never; 2: Seldom; 3: Sometimes; 4: Often; and 5:Very Often.

**1. Inspect the file**

What is the column separator in this file?

```python
# typically, network data will be collected aqs edgelists and saved as a file

# network data can be stored in many different formats!
# before creating a network data structure from a file, you need to check which format has been used.

a=open("data/cross_parker-consulting_info.txt","r")
for i in a:
    print (i)
a.close()

#what is the column separator in this file?

'''
The column separator in the file is determined by inspecting the printed lines. 
If the data is tab-separated, you will see \t between columns. 
If it's comma-separated, you will see , between columns, and so on.
'''
```

**2. Let's use networkx to create a network from data**    
First, import the networkx package  
Then create an empty graph  

```python
import networkx as nx
G=nx.Graph()

# The networkx package is imported, and an empty graph G is created.
```

**3. Create a Graph object (network) while reading the file**

```python
with open("data/Cross_Parker-Consulting_info.txt", "r") as f:
    for line in f:
        if line.startswith("#"):
            continue  # Skip comment lines
        parts = line.strip().split("\t")  # Assuming tab-separated
        if len(parts) == 3:
            node1, node2, weight = parts
            G.add_edge(node1, node2, weight=int(weight))

G

# A Graph object G is created by reading the file and adding edges with weights representing the frequency of information or advice requests.
```

**4. Inspect the Graph**  
How many nodes are there?  
How many edges are there?  

```python
num_nodes = G.number_of_nodes()
num_edges = G.number_of_edges()

print(f"Number of nodes: {num_nodes}")
print(f"Number of edges: {num_edges}")
```

**5. How do we store the information about the edges properties?**

```python
for edge in G.edges(data=True):
    print(edge)
```

**6. Let's plot the network**

```python
import matplotlib.pyplot as plt

pos = nx.spring_layout(G)  # Or any other layout
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=500, font_size=10)
plt.show()
```

**7. Let's plot the attributes of the edges**

```python
edges = G.edges(data=True)
weights = [edge[2]['weight'] for edge in edges]

nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=500, font_size=10, width=weights)
plt.show()
```
